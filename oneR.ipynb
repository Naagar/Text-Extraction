```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt

# Read Image
gray_im = cv2.imread('one.png')
gray_im = cv2.cvtColor(gray_im, cv2.COLOR_BGR2GRAY)
# Bluring 
gray_im = cv2.medianBlur(gray_im,5)
gray_im=cv2.GaussianBlur(gray_im, (3, 3), 0)

# Binary 
thresh = 127
gray_im = cv2.threshold(gray_im, thresh, 255, cv2.THRESH_BINARY)[1]
img  = gray_im
img_orignal  = gray_im
# Erosion

kernel1 = np.ones((5,5),np.uint8)
for i in range(0,10):
    gray_im = cv2.erode(gray_im,kernel1,iterations = 1)
    
# orignal minus eroded
img = img - gray_im

plt.subplot(111),plt.imshow(img_orignal, cmap = 'gray')
plt.title('I-1 '), plt.xticks([]), plt.yticks([])
plt.show()
plt.subplot(111),plt.imshow(gray_im, cmap = 'gray')
plt.title('I-1 '), plt.xticks([]), plt.yticks([])
plt.show()
plt.subplot(111),plt.imshow(img, cmap = 'gray')
plt.title('I-1 '), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
import cv2
import numpy as np

img = cv2.imread('one.png',0)
img = cv2.medianBlur(img,5)
cimg = cv2.cvtColor(img,cv2.COLOR_GRAY2BGR)

circles = cv2.HoughCircles(img,cv2.HOUGH_GRADIENT,1,10,
                            param1=50,param2=12,minRadius=0,maxRadius=20)

circles = np.uint16(np.around(circles))
for i in circles[0,:]:
    # draw the outer circle
    cv2.circle(cimg,(i[0],i[1]),i[2],(0,255,0),2)
    # draw the center of the circle
    cv2.circle(cimg,(i[0],i[1]),2,(0,0,255),3)

plt.subplot(111),plt.imshow(cimg, cmap = 'gray')
plt.title('I-1 '), plt.xticks([]), plt.yticks([])
plt.show()
```
