```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt

# Read image
img = cv2.imread('one.png')


# RGB to Gray
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Bluring image
for i in range(0,2):
    #print('skdjb')
    gray_image = cv2.medianBlur(gray_image,5)

# Gray to Binary
thresh = 127
im_bw = cv2.threshold(gray_image, thresh, 255, cv2.THRESH_BINARY)[1]

# Erosion
kernel = np.ones((5,5),np.uint8)
#for j in range(0,2):
#    im_bw = cv2.erode(im_bw,kernel,iterations = 1)
    
# Opening 
im_bw = cv2.morphologyEx(im_bw, cv2.MORPH_OPEN, kernel)
# Closing
im_bw = cv2.morphologyEx(im_bw, cv2.MORPH_CLOSE, kernel)


plt.subplot(111),plt.imshow(im_bw, cmap = 'gray')
plt.title('Result Image '), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
import cv2
import numpy as np;
im = cv2.imread('blob.jpg', cv2.IMREAD_GRAYSCALE)

thresh = 127
im = cv2.threshold(im, thresh, 255, cv2.THRESH_BINARY)[1]


# Setup SimpleBlobDetector parameters.
params = cv2.SimpleBlobDetector_Params()

# Change thresholds
params.minThreshold = 50
params.maxThreshold = 150
# Filter by Area.
params.filterByArea = True
params.minArea = 150
detector = cv2.SimpleBlobDetector_create(params)

# Detect blobs.
keypoints = detector.detect(im)

# Draw detected blobs as red circles.
# cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS ensures
# the size of the circle corresponds to the size of blob

im_with_keypoints = cv2.drawKeypoints(im, keypoints, np.array([]), (255,0,0), cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)



plt.subplot(111),plt.imshow(im_with_keypoints, cmap = 'gray')
plt.title('I-1 '), plt.xticks([]), plt.yticks([])
plt.show()
# Show blobs
#cv2.imshow("Keypoints", im_with_keypoints)
#cv2.waitKey(0)
#cv2.destroyAllWindows()
```

```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt


gray_im = cv2.imread('two.png')
gray_im = cv2.cvtColor(gray_im, cv2.COLOR_BGR2GRAY)
for i in range(0,5):
    gray_im = cv2.medianBlur(gray_im,5)

thresh = 127
gray_im = cv2.threshold(gray_im, thresh, 255, cv2.THRESH_BINARY)[1]
img  = gray_im
img_orignal  = gray_im
    
kernel = np.ones((5,5),np.uint8)
for i in range(0,10):
    gray_im = cv2.erode(gray_im,kernel,iterations = 1)
    
img = img - gray_im


plt.subplot(131),plt.imshow(img_orignal, cmap = 'gray')
plt.title('I-1 '), plt.xticks([]), plt.yticks([])
plt.show()
plt.subplot(132),plt.imshow(gray_im, cmap = 'gray')
plt.title('I-1 '), plt.xticks([]), plt.yticks([])
plt.show()
plt.subplot(133),plt.imshow(img, cmap = 'gray')
plt.title('I-1 '), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt


gray_im = cv2.imread('three.png')
gray_im = cv2.cvtColor(gray_im, cv2.COLOR_BGR2GRAY)
for i in range(0,5):
    gray_im = cv2.medianBlur(gray_im,5)

thresh = 127
gray_im = cv2.threshold(gray_im, thresh, 255, cv2.THRESH_BINARY)[1]
img  = gray_im
img_orignal  = gray_im
    
kernel = np.ones((5,5),np.uint8)
for i in range(0,10):
    gray_im = cv2.erode(gray_im,kernel,iterations = 1)
    
img = img - gray_im

fig = plt.figure()
a = fig.add_subplot(1, 3, 1)
imgplot = plt.imshow(img_orignal, cmap = 'gray')
a.set_title('img_orignal')

a = fig.add_subplot(1, 3, 2)
imgplot = plt.imshow(gray_im, cmap = 'gray')
imgplot.set_clim(0.0, 0.7)
a.set_title('Intermediat')

a = fig.add_subplot(1, 3, 3)
imgplot = plt.imshow(img, cmap = 'gray')
a.set_title('Final')



```
